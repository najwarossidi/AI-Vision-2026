```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

nama_pelajar = ["Ali", "Abu", "Siti", "Aina", "Kumar"]
markah = np.array([80, 75, 90, 85, 70])

print("=== ANALISIS MARKAH PELAJAR ===")
print("Semua Markah :", markah)
print("Markah Tertinggi :", np.max(markah))
print("Markah Terendah :", np.min(markah))
print("Purata Markah :", np.mean(markah))

print("\nSummary")
print("Bilangan Pelajar :", len(markah))
print("Jumlah Markah :", np.sum(markah))
print("Sisihan Piawai :", np.std(markah))

df = pd.DataFrame({
    "Nama Pelajar": nama_pelajar,
    "Markah": markah
})

print("\n=== DATAFRAME ===")
print(df)

plt.figure(figsize=(7,5))
plt.bar(df["Nama Pelajar"], df["Markah"])

plt.title("Prestasi Markah Pelajar")
plt.xlabel("Nama Pelajar")
plt.ylabel("Markah")
plt.ylim(0,100)

plt.show()
```
